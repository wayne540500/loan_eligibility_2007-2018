```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(context='notebook',style="ticks",palette="GnBu_d",font_scale=1.5,font='ETBembo',rc={"figure.figsize": (10, 6)})
import warnings
warnings.filterwarnings('ignore') 
pd.set_option('precision', 5) 
pd.set_option('display.float_format', lambda x: '%.4f' % x) 
pd.options.display.max_rows = 100 
```

```python
data=pd.read_csv("loan.csv")
data.shape
data.head()
```

```python
#只選擇2015年的數據
data_15=data[(data.issue_d=='Jan-2015')\
            |(data.issue_d=='Feb-2015')\
            |(data.issue_d=='Mar-2015')\
            |(data.issue_d=='Apr-2015')\
            |(data.issue_d=='May-2015')\
            |(data.issue_d=='Jun-2015')\
            |(data.issue_d=='Jul-2015')\
            |(data.issue_d=='Aug-2015')\
            |(data.issue_d=='Sep-2015')\
            |(data.issue_d=='Oct-2015')\
            |(data.issue_d=='Nov-2015')\
            |(data.issue_d=='Dec-2015')\
            ]
```

```python
#選擇需要分析的參數
analysis_columns=['issue_d','term','int_rate','grade','home_ownership','verification_status','purpose','loan_amnt','total_pymnt','out_prncp',
                  'total_rec_int','total_rec_prncp','installment','annual_inc','dti','fico_range_low','fico_range_high','last_fico_range_low',
                  'last_fico_range_high', 'open_acc','loan_status','delinq_amnt','acc_now_delinq','tot_coll_amt']
analysis_data = (data.loc[data.issue_d.notnull(),analysis_columns])\
                     .assign(month=lambda x: pd.to_datetime(x['issue_d']).dt.month,
                             interest_rate = lambda x:x['int_rate'],
                             fico_change = lambda x :  (x['last_fico_range_high']-x['fico_range_high']))
analysis_data.head()
```

```python
#對loan_amnt按月分類繪製貸款筆數總額圖
perform_data = analysis_data.groupby('month')['loan_amnt'].agg(['count','sum'])
f, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
x = perform_data.index
y1 = perform_data['count']
sns.barplot(x, y1, ax=ax1)
y2 = perform_data['sum']
ax1.set_xlabel("")
ax1.set_ylabel("loan_count")
sns.barplot(x, y2,ax=ax2)
ax2.set_ylabel("loan_amount")
sns.despine(bottom=True)
```

```python
#使用norm分布
from scipy.stats import norm
sns.distplot(analysis_data.loan_amnt,fit=norm,kde=False)
sns.despine(top=True)
```

```python
#繪製餅狀圖
analysis_data.term.value_counts().plot.pie(autopct='%.2f',figsize=(10, 10))
```

```python
#利率分布圖
sns.distplot(analysis_data.interest_rate.astype(float),fit=norm,kde=False)
sns.despine(top=True)
```

```python
#房屋拥有的情况
analysis_data.home_ownership.value_counts().plot.pie(autopct='%.2f',figsize=(10, 10))
```
